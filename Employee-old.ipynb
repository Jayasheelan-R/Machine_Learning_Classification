## Importing and altering the data

```python
import pandas as pd
```

```python
emp = pd.read_csv('Train.csv')
```

```python
emp.head()
```

```python
emp['Age'] = emp['Age'].fillna(0)
emp['Time_of_service'] = emp['Time_of_service'].fillna(0)
emp['Work_Life_balance'] = emp['Work_Life_balance'].fillna(0)
emp['Post_Level'] = emp['Post_Level'].fillna(0)
emp['Pay_Scale'] = emp['Pay_Scale'].fillna(0)
emp['Travel_Rate'] = emp['Travel_Rate'].fillna(0)
emp['growth_rate'] = emp['growth_rate'].fillna(0)
```

```python
emp['VAR1'] = emp['VAR1'].fillna(0)
emp['VAR2'] = emp['VAR2'].fillna(0)
emp['VAR3'] = emp['VAR3'].fillna(0)
emp['VAR4'] = emp['VAR4'].fillna(0)
emp['VAR5'] = emp['VAR5'].fillna(0)
emp['VAR6'] = emp['VAR6'].fillna(0)
emp['VAR7'] = emp['VAR7'].fillna(0)
```

## Train Test Split and some Modifications in data

```python
y_val = emp['Attrition_rate']
```

```python
x_data = emp.drop('Attrition_rate',axis = 1)
```

```python
x_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_val, test_size=0.2, random_state=3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = pd.DataFrame(data = scaler.transform(X_train),columns = X_train.columns, index = X_train.index)
```

```python
X_test = pd.DataFrame(data = scaler.transform(X_test),columns = X_test.columns, index = X_test.index)
```

## Training the Model

```python
import tensorflow as tf
```

```python
emp_id = tf.feature_column.numeric_column('Employee_ID')
gender = tf.feature_column.numeric_column('Gender')
age = tf.feature_column.numeric_column('Age')
education_lvl = tf.feature_column.numeric_column('Education_Level')
relationship = tf.feature_column.numeric_column('Relationship_Status')
hometown = tf.feature_column.numeric_column('Hometown')
unit = tf.feature_column.numeric_column('Unit')
decision_skill = tf.feature_column.numeric_column('Decision_skill_possess')
service = tf.feature_column.numeric_column('Time_of_service')
promotion = tf.feature_column.numeric_column('Time_since_promotion')
growth = tf.feature_column.numeric_column('growth_rate')
travel = tf.feature_column.numeric_column('Travel_Rate')
post = tf.feature_column.numeric_column('Post_Level')
pay = tf.feature_column.numeric_column('Pay_Scale')
benefits = tf.feature_column.numeric_column('Compensation_and_Benefits')
life = tf.feature_column.numeric_column('Work_Life_balance')
var1 = tf.feature_column.numeric_column('VAR1')
var2 = tf.feature_column.numeric_column('VAR2')
var3 = tf.feature_column.numeric_column('VAR3')
var4 = tf.feature_column.numeric_column('VAR4')
var5 = tf.feature_column.numeric_column('VAR5')
var6 = tf.feature_column.numeric_column('VAR6')
var7 = tf.feature_column.numeric_column('VAR7')
```

```python
feat_cols = [emp_id,gender,age,education_lvl,relationship,hometown,unit,decision_skill,service,promotion,
            growth,travel,post,pay,benefits,life,var1,var2,var3,var4,var5,var6,var7]
```

```python
lol = ['emp_id','gender','age','education_lvl','relationship','hometown','unit','decision_skill','service','promotion',
            'growth','travel','post','pay','benefits','life','var1','var2','var3','var4','var5','var6','var7','Attrition_rate']
```

```python
input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size = 10, num_epochs =10, shuffle = True, num_threads=5)
```

```python
model = tf.compat.v1.estimator.DNNClassifier(hidden_units = [23,23,23], feature_columns=feat_cols)
```

```python
model.train(input_fn=input_func, steps=12000)
```

```python
eval_input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size = 10, num_epochs =10, shuffle = False, num_threads=5)
```

```python
model.evaluate(eval_input_func)
```

## Prediction

```python
pred_input_fun = tf.compat.v1.estimator.inputs.pandas_input_fn(x = X_test, batch_size = 50,num_epochs = 1,shuffle = False)
```

```python
pred = model.predict(pred_input_fun)
```

```python
prediction = list(pred)
```

```python
prediction
```

```python
hello = []
for pred in prediction:
    hello.append(pred['logistic'])
```

## Mean squared Error

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,hello)
```
